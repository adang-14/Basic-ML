```python
import numpy as np
import math
```

```python
def sigmoid(z):
    return(1/(1+np.exp(-z)))

def softmax(z):
    return np.exp(z) / np.sum(np.exp(z), axis=0)
```

```python
nsamp = 15 # samples in batch
ni = 8 # number predictors per sample
nh = 5 # number of hidden units
```

(a) The hidden layer step (10a).

```python
X = np.random.uniform(0,1,(nsamp,ni))
WH = np.random.uniform(0,1,(nh,ni))
bH = np.random.uniform(0,1,(nh,1))
ZH = np.dot(X,WH.T) + bH.T
UH = sigmoid(ZH)
```

(b) The output layer step (10b) for binary classification with a sigmoid output (3).

```python
no = 1 # binary output
WO = np.random.uniform(0,1,(no,nh))
bO = np.random.uniform(0,1,(no,1))
ZO = np.dot(ZH,WO.T) + bO.T
UO = sigmoid(ZO)
UO
```

(c) The output layer step (10b) for K-class classification with a softmax output (5).

```python
no = 4 # softmax output
WO = np.random.uniform(0,1,(no,nh))
bO = np.random.uniform(0,1,(no,1))
ZO = np.dot(ZH,WO.T) + bO.T
UO = np.apply_along_axis(softmax, 1, ZO)
UO
```
